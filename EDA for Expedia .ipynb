```python
# Extract the year from the 'Week' column
df['Year'] = df['Week'].str[:4]
df['Year']
```

```python
# Filter the DataFrame to include only the rows where the 'Year' column is equal to '2016'
data_2016 = df[df['Year'] == '2016']
data_2016
```

```python
# Filter the DataFrame to include only the rows where the 'Year' column is equal to '2017'
data_2017 = df[df['Year'] == '2017']
data_2017
```

```python
# Group the 2016 data by 'Platform Type Name' and 'Super Region' 
# then calculate the sum of 'Net Gross Booking Value USD' for each group.
agg_2016 = data_2016[['Platform Type Name', 'Super Region', 'Net Gross Booking Value USD']].groupby(['Platform Type Name', 'Super Region'], as_index=False).sum()
```

```python
# Group the 2017 data by 'Platform Type Name' and 'Super Region', 
# then calculate the sum of 'Net Gross Booking Value USD' for each group.
agg_2017 = data_2017[['Platform Type Name', 'Super Region', 'Net Gross Booking Value USD']].groupby(['Platform Type Name', 'Super Region'], as_index=False).sum()
```

```python
# Add a new column 'Year' to the agg_2016 DataFrame and set its value to '2016'
# and the agg_2017 DataFrame and set its value to '2017'

agg_2016['Year'] = '2016'
agg_2017['Year'] = '2017'
```

```python
#Combine both year's Data
combined_data = pd.concat([agg_2016, agg_2017])
combined_data
```

```python
# Create a bar plot using seaborn to visualize Net Gross Booking Value by Super Region and Year
sns.barplot(x='Super Region', y='Net Gross Booking Value USD', hue='Year', data=combined_data)
plt.title('Net Gross Booking Value by Platform and Region')
plt.xlabel('Super Region')
plt.ylabel('Net Gross Booking Value USD')
plt.show()
```

```python
# Group the DataFrame 'df' by 'Year' and 'Platform Type Name', and calculate the sum of 'Net Gross Booking Value USD' for each group
agg_year_platform = df[['Year', 'Platform Type Name', 'Net Gross Booking Value USD']].groupby(['Year', 'Platform Type Name'], as_index=False).sum()
agg_year_platform
```

```python
# Create a new figure for the plot
plt.figure()
sns.barplot(x='Year', y='Net Gross Booking Value USD', hue='Platform Type Name', data=agg_year_platform)
plt.title('Net Gross Booking Value by Platform and Year')
plt.xlabel('Year')
plt.ylabel('Net Gross Booking Value USD')
plt.legend(title='Platform Type Name')
plt.show()
```
